# Data Preprocessing

## Cleaning 'year'

```python
# car_info['year'].unique()
```

```python
# Here we will remove trash values and make obj into integer data type
car_info = car_info[car_info['year'].str.isnumeric()]
car_info['year'] = car_info['year'].astype(int) # Only works when pure number wont even consider if ',' is present
car_info.info()
```

## Cleaning 'Price'

```python
# car_info['Price'].unique()
```

```python
# Have to remove 'Ask For Price' value and removing ',' in string

car_info = car_info[car_info['Price'] != 'Ask For Price']
car_info['Price'] = car_info['Price'].str.replace(',', '') # removing ',' to use .astype(int)
car_info['Price'] = car_info['Price'].astype(int)
car_info.info()
```

## Cleaning 'kms_driven'

```python
car_info['kms_driven'].unique()
```

```python
car_info = car_info[car_info['kms_driven'] != 'Petrol']
car_info['kms_driven'] = car_info['kms_driven'].str.replace(',', '') # removing ',' to use .astype(int)
car_info['kms_driven'] = car_info['kms_driven'].str.replace(' kms', '') # removing ' kms' to use .astype(int)
car_info['kms_driven'] = car_info['kms_driven'].astype(int)
car_info.info()
```

```python
car_info.head()
```

## Cleaning 'name'

```python
def three_names(obj):
  L = []
  count = 0
  for i in range(0, len(obj)):
    if count != 3:
      L.append(obj[i])
      count+=1
    else:
      break
  return " ".join(L)
```

```python
car_info['name'] = car_info['name'].apply(lambda x:x.split())
car_info['name'] = car_info['name'].apply(three_names)
car_info
```

## Droping all NULL value rows

```python
car_info.dropna(inplace= True)
car_info
```

## Rmoving outliers

```python
car_info = car_info[car_info['Price']<6000000]
car_info
```

## Resetting the index

```python
car_info =car_info.reset_index(drop=True)
car_info
```

# Test / Train Split

```python
X = car_info[['name','company','year','kms_driven','fuel_type']]
y = car_info['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2, random_state=20)
```

# Applying our model And also applying the encoding

Over all i think it is way better way of making the model than my old way

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
# OneHotEncoding
ohe=OneHotEncoder()
ohe.fit(X[['name','company','fuel_type']])
# transforming the column using OneHotEncoder
column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),
                                    remainder='passthrough')
# Choosing the model
lr=LinearRegression()
```

```python
# By this pipe line method we can apply all the changes at once
pipe=make_pipeline(column_trans,lr)
pipe.fit(X_train,y_train)
```

# Our Predictions and evaluations using R2 score

```python
y_pred=pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

Using K Fold Cross Validation

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = pipe, X = X_train, y = y_train, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```
